```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import os
import matplotlib.patches as mpatches
```

```python
ride_path = os.path.join("C:/", "Resources", "ride_data.csv")
ride_df = pd.read_csv(ride_path)
ride_df

city_path = os.path.join("C:/", "Resources", "city_data.csv")
city_df = pd.read_csv(city_path)
```

```python
merged_view = pd.merge(ride_df, city_df, on = "city", how = "outer")
merged_view.head()
```

```python
#group according to city
viewbycity = merged_view.groupby(["city"])
viewbycity.head()
```

```python
#isolate one value of Type per city given multiple repeating values of Type in merged, grouped df:

citytype = viewbycity["type"].unique()
citytype
```

```python
#iterate through groupby city Type data and generate a colors list given the 3 possible city types

colors = []

for type in citytype:
    if type == "Urban":
        colors.append("lightcoral")
    elif type == "Suburban":
        colors.append("skyblue")
    elif type == "Rural":
        colors.append("gold")

#print(colors)

colors_new = set(colors)
print(colors_new)
```

```python
#get total fares sum by city and a count of # of rides per city for average calc

totalfarebycity = viewbycity["fare"].sum()
countoffaresbycity = viewbycity["ride_id"].count()

#calc average fare per city
avgfarebycity = totalfarebycity / countoffaresbycity
avg_fare = round(avgfarebycity,2)

#calc total rides per city
total_rides = viewbycity["ride_id"].nunique()

#of drivers per city
drivers_total = viewbycity["driver_count"].unique().reset_index()["driver_count"].map(lambda x: x[0]).tolist()
```

```python
#create bubblechart, define items in chart, pass in a list of colors defined in above steps
plt.title("Pyber Ride Sharing Data (2016)")
plt.xlabel("Total Number of Rides (Per City)")
plt.ylabel("Average Fare ($)")

urban_patch = mpatches.Circle((0.5, 0.5), 1, facecolor='lightcoral',label="Urban")
suburban_patch = mpatches.Circle((0.5, 0.5), 1, facecolor='skyblue',label="Suburban")
rural_patch = mpatches.Circle((0.5, 0.5), 1, facecolor='gold',label="Rural")


plt.legend(handles=[urban_patch, suburban_patch, rural_patch])

plt.scatter(total_rides,avg_fare,s=drivers_total, marker="o", color = colors, alpha=.75)
plt.show()
```
